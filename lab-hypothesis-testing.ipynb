In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here

# Filter Dragon and Grass type Pokemon
dragon_df = df[df["Type 1"] == "Dragon"]
grass_df = df[df["Type 1"] == "Grass"]

# Perform a two-tailed t-test on HP
t_statistic, p_value = st.ttest_ind(dragon_df["HP"], grass_df["HP"], equal_var=False)  # Unequal variance

# Set significance level
alpha = 0.05

# Analyze the results
if p_value < alpha:
  print(f"We reject the null hypothesis (p-value = {p_value:.4f} < {alpha}).")
  print("On average, Dragon Pokemon have statistically significant differences in HP compared to Grass Pokemon.")
else:
  print(f"We fail to reject the null hypothesis (p-value = {p_value:.4f} >= {alpha}).")
  print("There is not enough evidence to conclude a statistically significant difference in average HP between Dragon and Grass Pokemon at the 5% significance level.")

# Additional analysis (optional)
print(f"Average HP (Dragon): {dragon_df['HP'].mean():.2f}")
print(f"Average HP (Grass): {grass_df['HP'].mean():.2f}")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here
# Separar los datos por tipo de Pokémon
legendary_stats = df[df['Legendary'] == True][['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
non_legendary_stats = df[df['Legendary'] == False][['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Realizar ANOVA para cada estadística
for stat in legendary_stats.columns:
    f_statistic, p_value = st.f_oneway(legendary_stats[stat], non_legendary_stats[stat])
    print(f"Estadística F para {stat}: {f_statistic:.4f}")
    print(f"Valor p para {stat}: {p_value:.4f}")
    if p_value < 0.05:
        print(f"Existe una diferencia significativa en {stat} entre Pokémon Legendarios y No Legendarios.")
    else:
        print(f"No se encontró una diferencia significativa en {stat} entre Pokémon Legendarios y No Legendarios.")
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.

```python

# Define school and hospital coordinates
school_coords = (-118, 37)
hospital_coords = (-122, 34)

def calculate_distance(point_coords, ref_coords):
  """Calculates the Euclidean distance between two points."""
  x1, y1 = point_coords
  x2, y2 = ref_coords
  return ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5

# Add columns for distance to school and hospital
df['distance_to_school'] = df[['longitude', 'latitude']].apply(
    lambda x: calculate_distance(x, school_coords), axis=1
)
df['distance_to_hospital'] = df[['longitude', 'latitude']].apply(
    lambda x: calculate_distance(x, hospital_coords), axis=1
)

# Define thresholds for proximity
close_threshold = 0.50  # House is considered close if distance < 0.50

# Filter houses close to school or hospital
close_to_school = df[df['distance_to_school'] < close_threshold]
close_to_hospital = df[df['distance_to_hospital'] < close_threshold]
far_from_both = df[~(df['distance_to_school'] < close_threshold) & ~(df['distance_to_hospital'] < close_threshold)]

# Test hypothesis: Houses close to school/hospital are more expensive
# Group by proximity (Close vs Far)
#close_prices = close_to_school['median_house_value'].append(close_to_hospital['median_house_value'])
close_prices = pd.concat([close_to_school['median_house_value'], close_to_hospital['median_house_value']])
far_prices = far_from_both['median_house_value']

# Perform t-test (assuming unequal variances)
t_statistic, p_value = st.ttest_ind(close_prices, far_prices, equal_var=False)

# Analyze results
if p_value < 0.05:
  print(f"We reject the null hypothesis (p-value = {p_value:.4f} < 0.05).")
  print("Houses close to a school or hospital have statistically significant differences in price compared to houses far from both.")
else:
  print(f"We fail to reject the null hypothesis (p-value = {p_value:.4f} >= 0.05).")
  print("There is not enough evidence to conclude a statistically significant difference in price between houses close to a school/hospital and those far from both at the 5% significance level.")

# Additional analysis (optional)
print(f"Average price (Close): {close_prices.mean():.2f}")
print(f"Average price (Far): {far_prices.mean():.2f}")
```
